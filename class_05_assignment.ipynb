```python
import pandas as pd
import numpy as np
import sklearn
import scipy
```

```python
data = {'price':[110,105,115,120,110,130,150,100,105]}
df = pd.DataFrame(data)
df
```

#normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['price_normalization']= scaler.fit_transform(df[['price']])
df
```

#Standard Scaler

```python
data = {'price':[110,105,115,120,110,130,150,100,105]}
df = pd.DataFrame(data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['price_StandardScaler']= scaler.fit_transform(df[['price']])
df
```

#Max Abs

```python
data = {'price':[110,105,115,120,110,130,150,100,105]}
df = pd.DataFrame(data)
```

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
df['price_MaxAbs']= scaler.fit_transform(df[['price']])
df
```

#Robust Scaler

```python
data = {'price':[110,105,115,120,110,130,150,100,105]}
df = pd.DataFrame(data)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
df['price_normalization']= scaler.fit_transform(df[['price']])
df
```

#Log Transform

```python
data = {'price':[110,105,115,120,110,130,150,100,105]}
df = pd.DataFrame(data)
```

```python
df['price_log_np'] = np.log1p(df[['price']])
df
```

```python
from sklearn.preprocessing import FunctionTransformer
ft = FunctionTransformer(np.log1p)
df['price_log_sk'] = ft.fit_transform(df[['price']])
df
```
